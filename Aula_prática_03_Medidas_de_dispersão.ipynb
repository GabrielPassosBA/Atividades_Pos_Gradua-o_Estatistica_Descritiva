```r
#Amostra0
a0=iris$Sepal.Length
a0
```

```r
#Tamanho das sepalas
length(a0)
```

```r
#Ordenado
sort(a0)
```

```r
#Amplitude amostral
h0= diff(range(a0))
h0
```

```r
#Variância amostral
var0=var(a0)
var0
```

```r
#Variância populacional
var(a0)*(length(a0)-1)/length(a0)
```

```r
#Desvio padrão amostral
sd0=sd(a0)
sqrt(var0)
```

```r
#Média -2 desvios e +2desvios
mean(a0)-2*sd0
mean(a0)+2*sd0
```

```r
#Coeficiente de variação amostral
cv0=sd0/mean(a0)*100
cv0
```

```r
#amostra1
a1=iris$Petal.Length
a1
```

```r
sort(a1)
```

```r
#Amplitude amostral
h1= diff(range(a1))
h1
```

```r
#Variância amostral
var1=var(a1)
var1
```

```r
#Desvio padrão amostral
sd1=sd(a1)
sd1
```

```r
#Média -2 desvios e +2desvios
mean(a1)-2*sd1
mean(a1)+2*sd1
```

```r
#Coeficiente de variação amostral
cv1=sd(a1)/mean(a1)*100
cv1
```

```r
dados_a0_consolidados = data.frame ( "amplitude"=h0,
                                     "variancia"=var0,
                                     "desvio_padrao"=sd0,
                                     "media"=mean(a0),
                                     "coeficente_variacao"=cv0 )
dados_a0_consolidados
```

```r
dados_a1_consolidados = data.frame ( "amplitude"=h1,
                                     "variancia"=var1,
                                     "desvio_padrao"=sd1,
                                     "media"=mean(a1),
                                     "coeficente_variacao"=cv1 )
dados_a1_consolidados
```

```r
consolidado_comparativo = data.frame(t(dados_a0_consolidados),t(dados_a1_consolidados))

colnames(consolidado_comparativo)=c("resumo_tamanhho_Sepala","resumo_tamanhho_Petala")

consolidado_comparativo
```

```r
# Comparação
par(mfrow=c(1,2))
plot(a0,ylim =c(1,7),col=iris$Species)
abline(h=mean(a0))
abline(h=mean(a0)-2*sd0)
abline(h=mean(a0)+2*sd0)

plot(a1,col=iris$Species)
abline(h=mean(a1))
abline(h=mean(a1)-1.5*sd1)
abline(h=mean(a1)+1.5*sd1)
```

```r
# Comparação
par(mfrow=c(1,2))
plot(a0,ylim =c(1,7),col=iris$Species)
abline(h=quantile(a0,.50))
abline(h=quantile(a0,.25))
abline(h=quantile(a0,.75))

plot(a1,col=iris$Species)
abline(h=quantile(a0,.50))
abline(h=quantile(a1,.25))
abline(h=quantile(a1,.75))
```

```r
#Pela Ótica Tidyverse
dados_sepala = iris %>% summarise ( media0=mean(Sepal.Length),
                                   h0= diff(range(Sepal.Length)),
                                   var0=var(Sepal.Length),
                                   sd0=sd(Sepal.Length),
                                   cv0=sd0/media0*100
)

dados_sepala
```

```r
dados_petala = iris %>% summarise ( media1=mean(Petal.Length),
                                   h1= diff(range(Sepal.Length)),
                                   var1=var(Petal.Length),
                                   sd1=sd(Petal.Length),
                                   cv1=sd1/media1*100
)
dados_petala
```
